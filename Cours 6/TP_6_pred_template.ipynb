```python

# From Deep Learning for Time Series Forecasting (c) Jason Brownlee

#split a univariate sequence into samples
def split_sequence_univariate(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
            
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)


# Transforms a 2d-X and y to 3d X and y


# split a multivariate sequence into samples
def split_sequences_multivariate(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
# find the end of this pattern
        end_ix = i + n_steps
# check if we are beyond the dataset
        if end_ix > len(sequences)-1:
            break
# gather input and output parts of the pattern
        seq_x, seq_y = sequences[i:end_ix, :], sequences[end_ix, -1]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
def build_LSTM(neurons,X_shape,stacked=False,use_bias=True):
    model = Sequential()
    model.add(LSTM(neurons,input_shape=X_shape,
                   return_sequences=stacked,use_bias=use_bias))
    if stacked:
        model.add(LSTM(neurons,input_shape=X_shape))
        
        
    
### add some dense layers
    model.add(Dense(neurons // 2, activation='relu'))
    model.add(Dense(neurons // 4, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

### choose what kind of prediction you wish to learn

#    model.compile(loss='mean_squared_error', optimizer='adam')  
    model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
    return(model)

```

```python
batch_size=64
num_epochs=100

N_I=10   # number of relevant timesteps to make a prediction
dim_h_LSTM=10  # dimension of h_t
stacked=False

dT=5               # time between calibrations
dT_orig=dT

t1s=np.arange(batch_size+N_I,XY.values.shape[0]-dT,dT)
t1s
```

```python

myLSTM =   build_LSTM(dim_h_LSTM,(N_I,XY.shape[1]),stacked=stacked)
```

```python
allpreds=[]
dT=dT_orig

```

```python
for t1 in t1s:
    T_in = batch_size + N_I
    t0=t1-T_in
    print(t0,t1)
    X,y=split_sequences_multivariate(XY.values[(t1-batch_size-N_I):(t1),:],N_I)
    
    y = np.where(y < 0, 0 , 1)
    #wait: do you need to binarize y?

    #cold training
    myLSTM = build_LSTM(dim_h_LSTM,(N_I,XY.shape[1]),stacked=stacked)
    myLSTM.fit(X,y,
               batch_size=batch_size,epochs=num_epochs,
              callbacks=[es],validation_split=0.2,verbose=0)
    
    if XY.shape[0]<t1+dT:
        dT=XY.shape[0]-t1-1
    
    X_oos,y_oos=split_sequences_multivariate(XY.values[(t1-N_I):(t1+dT),:],N_I)
    y_oos = np.where(y_oos < 0, 0 , 1)
    

    preds=myLSTM.predict(X_oos,batch_size=dT)
    preds=preds.flatten()
  
    predsDF=pd.DataFrame({'time':np.arange(t1,t1+dT,1), 'preds':preds, 'real':y_oos})
    allpreds.append(predsDF)
    
dT=dT_orig
```
